## Step 6.5 (Alternative): ü§ñ Automated Batch Processing

**Auto-monitor Google Drive folders and process videos automatically!**

This replaces Steps 7-10 with a fully automated workflow:
- Drop videos into `input` folder in Google Drive
- Script automatically processes them with LAMA (fast)
- Outputs saved to `output` folder
- Processed inputs moved to `processed` folder

Perfect for continuous background processing!

````python
#@title üöÄ Launch Auto-Processing { display-mode: "form" }

#@markdown ---
#@markdown ### ‚öôÔ∏è Configuration
enable_auto_processing = True  #@param {type:"boolean"}
check_interval_seconds = 10    #@param {type:"slider", min:5, max:60, step:5}
drive_base_folder = "Figuro/sora-watermark-remover" #@param {type:"string"}
use_lama = True                #@param {type:"boolean"}
extra_output_path = "Figuro/video-upscale/input"         #@param {type:"string"}

import torch, time, os, sys, shutil, threading
from pathlib import Path
from datetime import datetime
from demark_world.core import DeMarkWorld
from demark_world.schemas import CleanerType
from google.colab import drive

# ‚îÄ‚îÄ Kill any previous monitor loop ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
_STOP_EVENT = getattr(sys.modules[__name__], '_STOP_EVENT', None)
if _STOP_EVENT is not None:
    print("‚èπÔ∏è  Stopping previous monitor...")
    _STOP_EVENT.set()
    time.sleep(2)
_STOP_EVENT = threading.Event()
sys.modules[__name__]._STOP_EVENT = _STOP_EVENT

# ‚îÄ‚îÄ 1. Drive (already mounted in Step 1) ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
if not os.path.isdir('/content/drive/MyDrive') and not os.path.isdir('/content/drive/Shareddrives'):
    raise RuntimeError("‚ùå Drive not mounted ‚Äî run Step 1 first.")
print("‚úÖ Drive already mounted")

# ‚îÄ‚îÄ 2. Auto-detect correct Shared Drive path ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
def find_drive_base(subfolder: str) -> str:
    candidates = [
        f"/content/drive/Shareddrives/{subfolder}",
        f"/content/drive/Shared drives/{subfolder}",
        f"/content/drive/MyDrive/{subfolder}",
    ]
    drive_root = "/content/drive"
    try:
        for top in os.listdir(drive_root):
            top_path = os.path.join(drive_root, top)
            if os.path.isdir(top_path):
                candidate = os.path.join(top_path, subfolder)
                if candidate not in candidates:
                    candidates.append(candidate)
    except Exception:
        pass

    print("\nüîç Searching for drive path...")
    for c in candidates:
        parent = str(Path(c).parent)
        if os.path.isdir(parent):
            print(f"  ‚úÖ Found: {c}")
            return c
        else:
            print(f"  ‚úó  {c}")

    print("\nüìÅ Available paths under /content/drive:")
    try:
        for top in os.listdir(drive_root):
            top_path = os.path.join(drive_root, top)
            print(f"  {top}/")
            try:
                for sub in os.listdir(top_path)[:5]:
                    print(f"    {sub}/")
            except Exception:
                pass
    except Exception as e:
        print(f"  Could not list: {e}")

    raise RuntimeError(
        f"Could not find '{subfolder}' under /content/drive.\n"
        "Check the drive_base_folder setting and that the Shared Drive is accessible."
    )

DRIVE_BASE = find_drive_base(drive_base_folder)
folders = {
    "input":     f"{DRIVE_BASE}/input",
    "processed": f"{DRIVE_BASE}/processed",
    "output":    f"{DRIVE_BASE}/output",
}
for p in folders.values():
    os.makedirs(p, exist_ok=True)

# ‚îÄ‚îÄ Extra output folder ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
_extra = extra_output_path.strip()
if _extra:
    # If user gave a bare subfolder (no /content/drive prefix), auto-resolve it
    if not _extra.startswith('/'):
        EXTRA_DIR = str(Path(DRIVE_BASE).parent.parent / _extra)
    else:
        EXTRA_DIR = _extra
    os.makedirs(EXTRA_DIR, exist_ok=True)
else:
    EXTRA_DIR = None

print("\n" + "=" * 60)
print("üìÇ FOLDER STRUCTURE READY")
print("=" * 60)
for k, v in folders.items():
    print(f"  {k:<10} {v}")
if EXTRA_DIR:
    print(f"  {'extra':<10} {EXTRA_DIR}")
print("=" * 60)

# ‚îÄ‚îÄ 3. GPU / CPU info ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
print("\nüñ•Ô∏è  COMPUTE INFO")
print("=" * 40)
if torch.cuda.is_available():
    gpu       = torch.cuda.get_device_properties(0)
    mem_total = gpu.total_memory / 1024**3
    mem_used  = torch.cuda.memory_allocated(0) / 1024**3
    print(f"  ‚úÖ GPU : {gpu.name}")
    print(f"     VRAM: {mem_used:.1f} GB used / {mem_total:.1f} GB total")
    print(f"     CUDA: {torch.version.cuda}")
else:
    print("  ‚ö†Ô∏è  No GPU ‚Äî running on CPU (will be slow!)")
print("=" * 40)

# ‚îÄ‚îÄ 4. Model init ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
cleaner_type = CleanerType.LAMA if use_lama else CleanerType.E2FGVI_HQ
model_label  = "LAMA" if use_lama else "E2FGVI_HQ"
print(f"\nüîß Initialising {model_label}‚Ä¶", flush=True)
demark = DeMarkWorld(cleaner_type=cleaner_type)
print(f"‚úÖ {model_label} ready!\n", flush=True)

# ‚îÄ‚îÄ 5. Helpers ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
SUPPORTED   = ('.mp4', '.mov', '.avi', '.mkv', '.webm')
_PROCESSING = set()
_PROC_LOCK  = threading.Lock()

def scan_input() -> list[str]:
    try:
        os.system("sync")
        all_files = sorted(
            f for f in os.listdir(folders["input"])
            if f.lower().endswith(SUPPORTED)
        )
        with _PROC_LOCK:
            return [f for f in all_files if f not in _PROCESSING]
    except FileNotFoundError:
        print("‚ö†Ô∏è  Input folder missing ‚Äî check Drive mount", flush=True)
        return []
    except Exception as e:
        print(f"‚ö†Ô∏è  Scan error: {e}", flush=True)
        return []

def process_file(filename: str):
    with _PROC_LOCK:
        if filename in _PROCESSING:
            return
        _PROCESSING.add(filename)

    input_path  = Path(folders["input"]) / filename
    ts          = datetime.now().strftime("%Y%m%d_%H%M%S")

    safe_stem   = Path(filename).stem.replace(" ", "_").replace("(","").replace(")","")
    safe_name   = f"{safe_stem}{Path(filename).suffix}"
    temp_path   = Path("/tmp") / safe_name
    out_name    = f"{ts}_{safe_stem}_cleaned.mp4"
    output_path = Path(folders["output"]) / out_name

    print(f"\n{'‚îÄ'*60}", flush=True)
    print(f"‚ñ∂  {filename}", flush=True)
    print(f"   ‚Üí {out_name}", flush=True)
    print(f"   Started : {datetime.now().strftime('%H:%M:%S')}", flush=True)
    print(f"{'‚îÄ'*60}", flush=True)

    t0 = time.time()
    try:
        shutil.copy2(str(input_path), str(temp_path))

        import tqdm as tqdm_module
        orig_file = getattr(tqdm_module.tqdm, 'file', None)
        tqdm_module.tqdm.file = sys.stdout

        demark.run(temp_path, output_path)

        if orig_file:
            tqdm_module.tqdm.file = orig_file

        temp_path.unlink(missing_ok=True)

        mins, secs = divmod(int(time.time() - t0), 60)
        print(f"\n‚úÖ Done in {mins}m {secs}s  ‚Üí  {out_name}", flush=True)

        # ‚îÄ‚îÄ Copy to extra output if configured ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
        if EXTRA_DIR:
            extra_path = Path(EXTRA_DIR) / out_name
            shutil.copy2(str(output_path), str(extra_path))
            print(f"üì§ Extra copy ‚Üí  {extra_path}", flush=True)

        archived = Path(folders["processed"]) / f"{ts}_{safe_name}"
        shutil.move(str(input_path), str(archived))
        print(f"üì¶ Archived  ‚Üí  processed/{archived.name}\n", flush=True)

    except Exception as e:
        mins, secs = divmod(int(time.time() - t0), 60)
        print(f"\n‚ùå Failed after {mins}m {secs}s: {e}", flush=True)
        temp_path.unlink(missing_ok=True)
        err_path = Path(folders["processed"]) / f"{ts}_ERROR_{safe_name}"
        try:
            shutil.move(str(input_path), str(err_path))
            print(f"üóÇÔ∏è  Moved to  processed/{err_path.name}\n", flush=True)
        except Exception as mv_err:
            print(f"   (could not move: {mv_err})\n", flush=True)
    finally:
        with _PROC_LOCK:
            _PROCESSING.discard(filename)

# ‚îÄ‚îÄ 6. Main loop (foreground) ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
if not enable_auto_processing:
    print("‚è∏Ô∏è  Disabled. Enable in configuration above.")
else:
    print(f"ü§ñ Watching  {folders['input']}")
    print(f"   Interval : {check_interval_seconds}s  |  Model: {model_label}")
    if EXTRA_DIR:
        print(f"   Extra out: {EXTRA_DIR}")
    print("   Interrupt kernel (‚ñ†) to stop.\n", flush=True)

    heartbeat = 0
    while not _STOP_EVENT.is_set():
        try:
            pending = scan_input()

            if pending:
                print(f"\nüé¨ {len(pending)} queued: {', '.join(pending)}", flush=True)
                for fn in pending:
                    if _STOP_EVENT.is_set():
                        break
                    process_file(fn)
                print("‚úî  Queue empty ‚Äî resuming watch‚Ä¶\n", flush=True)
                heartbeat = 0
            else:
                heartbeat += 1
                print("¬∑", end=("\n" if heartbeat % 30 == 0 else ""), flush=True)

            time.sleep(check_interval_seconds)

        except KeyboardInterrupt:
            print("\n\nüõë Stopped by user.")
            break
        except Exception as e:
            print(f"\n‚ö†Ô∏è  Loop error: {e} ‚Äî retrying in 30s‚Ä¶", flush=True)
            time.sleep(30)
````

````python
# Process all videos in batch
from pathlib import Path
from demark_world.core import DeMarkWorld
from demark_world.schemas import CleanerType
from tqdm import tqdm

batch_input = Path("batch_input")
batch_output = Path("batch_output")

# Get all video files
video_files = list(batch_input.glob("*.mp4")) + list(batch_input.glob("*.mov")) + list(batch_input.glob("*.avi"))

print(f"Processing {len(video_files)} videos...")
print("")

# Initialize model (LAMA for speed)
demark = DeMarkWorld(cleaner_type=CleanerType.LAMA)

# Process each video
for video_file in tqdm(video_files, desc="Processing videos"):
    output_file = batch_output / f"cleaned_{video_file.name}"
    print(f"\nProcessing: {video_file.name}")

    try:
        demark.run(video_file, output_file)
        print(f"‚úÖ Completed: {output_file.name}")
    except Exception as e:
        print(f"‚ùå Error processing {video_file.name}: {e}")

print("\n" + "="*60)
print("‚úÖ Batch processing complete!")
print("="*60)
````

````python
# Download all processed videos
from google.colab import files
from pathlib import Path

batch_output = Path("batch_output")

print("Downloading all processed videos...")
for video_file in batch_output.glob("*.mp4"):
    print(f"Downloading: {video_file.name}")
    files.download(str(video_file))

print("\n‚úÖ All downloads started!")
````
